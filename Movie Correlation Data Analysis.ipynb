```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# read in the data
dataFrame = pd.read_csv(r'C:\Users\Ange Lousi\Desktop\movies.csv')
```

```python
dataFrame.head()
```

```python
dataFrame.dtypes
```

```python
# This will drop any rows with null values
dataFrame = dataFrame.dropna(how='any',axis=0) 
```

```python
dataFrame.head()
```

```python
#correcting the year table as the year was inconsistet  with the release year

dataFrame['yearReleased'] = dataFrame['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
dataFrame = dataFrame.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#look at all the data
pd.set_option('display.max_rows', None)
dataFrame.head(10)
```

```python
#drop the duplicates of entire dataFrame
# dataFrame.drop_duplicates().sort_values(ascending=False)

#drop dulicates from the 'company' col
# dataFrame['company'] = dataFrame['company'].drop_duplicates()

# dataFrame['company'].drop_duplicates().sort_values(ascending=False)

dataFrame.drop_duplicates()
dataFrame.head(10)
```

```python
#My predictions: the budget correlation with the gross revenue will have a high correlation
#As well as the Company having a  high correlation with the gross revenue

#to compare the two and see the results really quickly is to make a scatter plot
```

```python
#just sorting dataFrame by gross
dataFrame = dataFrame.sort_values(by=['gross'], inplace=False, ascending=False)

#scatter plot Budget vs. Gross
plt.scatter(x=dataFrame['budget'], y=dataFrame['gross'])
#just gonna add on more info, so the graph is more readable
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
#to view scatter plot
plt.show()
```

```python
#regression plot to plot the budget vs. Gross
#lets add extra info to look nicer: scatter_kws to change color of dots and color of line
sns.regplot(x='budget', y='gross', data=dataFrame, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#lets start looking at the correlation
#an easy way to look is df.corr() that ONLY uses numerical fields

dataFrame.corr(method='pearson') #pearson, kendall, spearman
```

```python
#visualise the table above^^ and make a correlation matrix 
correlation_matrix = dataFrame.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
#just gonna add on more info, so the graph is more readable
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#so looking at the matrix above, the darker the color the lower the correlation, the lighter the color, the higher the correlation
dataFrame.head()
```

```python
#Now were gonna look at the rest of the columns that are NOT numeric and try to create a numeric feature for it

dataFrameNumerized = dataFrame

for col in dataFrameNumerized.columns:
    if(dataFrameNumerized[col].dtype == 'object'):
        dataFrameNumerized[col] = dataFrameNumerized[col].astype('category')
        dataFrameNumerized[col] = dataFrameNumerized[col].cat.codes

dataFrameNumerized.head()
```

```python
dataFrame.head()
```

```python
#visualise the now "numerized" table above^^ and make a new correlation matrix 

correlation_dataFrameNumerized_matrix = dataFrameNumerized.corr(method='pearson')
sns.heatmap(correlation_dataFrameNumerized_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
dataFrameNumerized.corr()
```

```python
#unstaking the table above

correlation_mat = dataFrameNumerized.corr()
correlation_pairs = correlation_mat.unstack()

correlation_pairs.head(30)
```

```python
#sorting the correlation pairs
sorted_corrleation_pairs = correlation_pairs.sort_values()

sorted_corrleation_pairs.head(30)
```

```python
#filtering the sorted pairs by high correlation
highCorrelation = sorted_corrleation_pairs[(sorted_corrleation_pairs) >= 0.5 ]

highCorrelation
```
